```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
Bangalore=['koramangala_bangalore','marathahalli_bangalore','hsr_layout_bangalore',
        'whitefield_bangalore','indira_nagar_bangalore','bellandur_bangalore','chandra_layout_bangalore',
        'j_p_nagar_bangalore','btm_layout_bangalore','jayanagar_bangalore']
```

```python
url=[]
for i in Bangalore:
    url.append('https://www.nobroker.in/flats-for-sale-in-'+i)
```

```python
Gname=[]
Grate=[]
Gemi=[]
Garea=[]
Glocation=[]
for i in range(len(url)):
    data=requests.get(url[i])
    soup=BeautifulSoup(data.text,'html.parser')
    locname=soup.find_all('a',class_='overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full')#name
    locrate=soup.find_all('div','font-semi-bold heading-6')#rate
    #Name
    Name=[]                       
    for j in locname: 
        Name.append(j.text)      
    Gname.append(Name)
    #Rate
    rate=[]
    lorate=[]
    for j in locrate:
        rate.append(j.text)
    for j in range(0,len(rate),3):
        lorate.append(rate[j])
    Grate.append(lorate)
    #EMI
    EMI=[]
    for j in range(1,len(rate),3):
        EMI.append(rate[j])
    Gemi.append(EMI)
    #Area
    area=[]
    for j in range(2,len(rate),3):
        area.append(rate[j])
    Garea.append(area)
    #Location
    location=[]
    for j in range(len(Name)):
        location.append(Bangalore[i])
    Glocation.append(location)
# len(Gname)
# len(Grate)
# len(Gemi)
# len(Garea)
```

```python
#BHK
Flat_Type=[]
for i in Gname:
    for j in i:
        Flat_Type.append(j)
# print(Flat_Type)
# Price
Price=[]
for i in Grate:
    for j in i:
        Price.append(j)
# print(Price)
# EMI
Monthly_Installment=[]
for i in Gemi:
    for j in i:
        Monthly_Installment.append(j)
# print(Monthly_Installment)
# Area
Build_Area=[]
for i in Garea:
    for j in i:
        Build_Area.append(j)
# print(Build_Area)
#Location
Loc=[]
for i in Glocation:
    for j in i:
        Loc.append(j)
        
```

```python
df=pd.DataFrame({'Location':Loc,'Flat Type':Flat_Type,'Price':Price,'EMI':Monthly_Installment,'Area':Build_Area})
```
